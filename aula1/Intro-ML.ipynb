## Dividindo a base em treino e teste

Utilizar a própria base de treinamento para avaliar o desempenho de um algoritmo pode causar o problema conhecido como "overfitting".

Para aliviar esse problema, é possível dividir o conjunto de treinamento entre "teste" e "treino". Para ajudar neste processo, existe a função: train_test_split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) 

print("X ",X_train.shape, X_test.shape, " y ", y_train.shape, y_test.shape)
print(X_train.head())
print(y_train.head())
```

## Scaling

A escala, ou grandeza, dos valores de cada variável pode atrapalhar alguns algoritmos de classificação. Para aliviar esse problema, é possível padronizar a escala para que não haja diferenças significativas de escala.

 Exemplo: peso e altura

### Min Max: define um range para que os dados sejam re-escalonados

```python
# data values
X = [ [110, 200], [120, 800], [310, 400], [140, 900], [510, 200], [653, 400] ,[310, 880] ]

# transform data
scaler = MinMaxScaler(feature_range=(0, 3))
rescaledX = scaler.fit_transform(X)

# summarize transformed data
np.set_printoptions(precision=2)
print(rescaledX)
#print(rescaledX[0:6,:])
```

### Standandard: range de valores é definida com base no intervalo da base

```python
# data values
X = [ [110, 200], [120, 800], [310, 400], [140, 900], [510, 200], [653, 400] ,[310, 880] ]

# scaler
scaler = StandardScaler().fit(X)
rescaledX = scaler.transform(X)

# summarize transformed data
np.set_printoptions(precision=3)
print(rescaledX[0:6,:])
```

## Imputer

Em alguns casos, algumas bases de dados contém algumas linhas com dados faltantes, seja por alguma falha na coleta ou por alguma falha no armazenamento. Para estes casos, é necessário substituir essas lacunas por um valor real. Existem diversas estratégias para se fazer isso, e suas correspondentes implementações disponíveis.

### Substitui dados faltantes
* strategy
* mean
* median
* most_frequent
* constant

```python
imp = SimpleImputer(missing_values=np.nan, strategy='median')

X = [[np.nan, 2], [6, np.nan], [7, 6]]

imp.fit(X)
x_sem_null=imp.transform(X)

print(x_sem_null)
```
